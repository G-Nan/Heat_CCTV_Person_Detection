```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
colors = sns.color_palette("pastel")
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('Dataset/사관학교별 경쟁률.csv', encoding = 'cp949')
df = df[df['연도'] >= 2016]
df.head(10)
```

```python
list_school = df['구분'].unique()

i = 0

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

for s in list_school:
    
    
    df_s = df[df['구분'] == s]
    
    if i != 3:
        sns.lineplot(x = df_s['연도'], y = df_s['전체 경쟁률'], color = colors[i], ax = ax1, label = s, linewidth = 2)
        
    else:
        sns.lineplot(x = df_s['연도'], y = df_s['전체 경쟁률'], color = colors[i], ax = ax2, label = s, linewidth = 2)
        ax2.legend(bbox_to_anchor=(1, 0.75)) 
        ax2.grid(False)
       
    i += 1
    
plt.title('사관학교별 경쟁률', fontsize = 20)
plt.show()
```

```python
s = list_school[1]
df_s = df[df['구분'] == s]
sns.lineplot(x = df_s['연도'], y = df_s['전체 경쟁률'], color = colors[i], linestyle = '-')
sns.lineplot(x = df_s['연도'], y = df_s['남자 경쟁률'], color = colors[i], linestyle = ':')
sns.lineplot(x = df_s['연도'], y = df_s['여자 경쟁률'], color = colors[i], linestyle = '--')
print(s)
```
