```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
colors = sns.color_palette("pastel")
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('Dataset/south-korea-military-spending-defense-budget.csv', header = 9)
df
```

```python
df = df.iloc[45:-1, :]
df
```

```python
for row in range(len(df)):
    df.iloc[row, 0] = df.iloc[row, 0].split('-')[0]
```

```python
plt.figure(figsize = (16, 4))
sns.lineplot(x = df['date'], y = df[df.columns[1]], color = colors[4], linewidth = 3)
plt.title('연도별 국방 예산', fontsize = 25)
plt.xlabel('연도')
plt.ylabel('국방 예산 ($)')
plt.show()
```

```python
df = pd.read_csv('Dataset/국방예산 추이.csv', encoding = 'cp949')
df
```

```python
for row in range(len(df)):
    df.iloc[row, 1] = df.iloc[row, 1] / 10000
df
```

```python
plt.figure(figsize = (16, 4))
sns.lineplot(x = df['연도'], y = df['국방비'], color = colors[4], linewidth = 3)
plt.title('연도별 국방 예산', fontsize = 25)
plt.xlabel('연도')
plt.ylabel('국방 예산 (조)')
plt.show()
```
