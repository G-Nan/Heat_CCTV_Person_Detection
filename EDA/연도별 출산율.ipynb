```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
colors = sns.color_palette("pastel")
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('Dataset/연도별 출산율.csv', encoding = 'cp949')
df
```

```python
plt.figure(figsize = (16, 4))
sns.lineplot(x = df['연도'], y = df['합계 출산율'], color = colors[5], linewidth = 3)
plt.title('연도별 출산율', fontsize = 25)
plt.xlabel('연도')
plt.ylabel('출산율 (명)')
plt.show()
```
