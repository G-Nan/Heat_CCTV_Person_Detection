```python
import os
import numpy as np
import pandas as pd
import pydeck as pdk
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
colors = sns.color_palette("pastel")
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
list_df = []

list_name = os.listdir('Dataset/현역병_모집실적')
for name in list_name:
    year = name.split('(')[1][:4]
    mil_type = name.split(' ')[0][:-1]
    
    path = f'Dataset/현역병_모집실적/{name}'
    df = pd.read_csv(path, encoding = 'cp949')
    df['연도'] = year
    df['종류'] = mil_type
    list_df.append(df)
    
df = pd.concat(list_df, axis = 0).reset_index(drop = True)
df
```

```python
for row in range(len(df)):
    df.iloc[row, 7] = int(df.iloc[row, 7].replace(',', ''))
    
df
```

```python
df_g = pd.DataFrame(df.groupby(['구분', '연도'])['실입영'].sum()).reset_index()
df_g
```

```python
list_year = df['연도'].unique()
list_city = df['구분'].unique()

for city in list_city:
    df_c = df_g[df_g['구분'] == city]
    sns.lineplot(x = df_c['연도'], y = df_c['실입영'])
```

```python
import plotly.graph_objects as go
import pandas as pd

# 예제 데이터 준비
data = {
    'City': ['Seoul', 'Daegu', 'Busan', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan', 'Sejong', 'Gyeonggi', 'Gangwon', 'Chungbuk', 'Chungnam', 'Jeonbuk', 'Jeonnam', 'Gyeongbuk', 'Gyeongnam', 'Jeju'],
    'Latitude': [37.5665, 35.8722, 35.1796, 37.4563, 35.1595, 36.3510, 35.5384, 36.4874, 37.4138, 37.8228, 36.6357, 36.6589, 35.7175, 34.8679, 36.4919, 35.2383, 33.4996],
    'Longitude': [126.9780, 128.6018, 129.0756, 126.7052, 126.8526, 127.3850, 129.3114, 127.2825, 127.5183, 128.1555, 127.4914, 126.7052, 127.1530, 127.2822, 128.8889, 128.6925, 126.5312],
    'Value': [9776000, 2437000, 3449000, 2957000, 1505000, 1538000, 1145000, 342000, 12690000, 1567000, 1602000, 2125000, 1808000, 1863000, 2663000, 3347000, 674000]
}

df = pd.DataFrame(data)

# 지도 생성
fig = go.Figure()

# 각 도시에 막대 추가
for i in range(len(df)):
    fig.add_trace(go.Scattermapbox(
        lat=[df['Latitude'][i]],
        lon=[df['Longitude'][i]],
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=12,
            color='rgb(0, 255, 0)',
            opacity=0.7
        ),
        text=df['City'][i],
        hoverinfo='text'
    ))

# 3D 막대 추가
fig.add_trace(go.Bar3d(
    x=df['Longitude'],
    y=df['Latitude'],
    z=[0]*len(df),
    dx=[0.1]*len(df),
    dy=[0.1]*len(df),
    dz=df['Value'],
    color='green',
    opacity=0.7
))

# 레이아웃 설정
fig.update_layout(
    title='South Korean Cities 3D Bar Graph',
    autosize=True,
    hovermode='closest',
    mapbox=dict(
        accesstoken='your_mapbox_access_token',  # Mapbox Access Token 필요
        bearing=0,
        center=dict(
            lat=36.5,
            lon=127.5
        ),
        pitch=0,
        zoom=6,
        style='carto-darkmatter'
    ),
    scene=dict(
        xaxis_title='Longitude',
        yaxis_title='Latitude',
        zaxis_title='Value'
    )
)

fig.show()
```
