```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
colors = sns.color_palette("pastel")
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('Dataset/병무청_현역병(징집병_모집병) 입영현황_20221231.csv', encoding = 'CP949')
df['총합'] = df['현역(징집)병 입영인원'] + df['모집병 입영인원']
df = df.loc[df['연도'] > 2010, :]
df
```

```python
plt.figure(figsize = (6, 6))
sns.lineplot(x = df['연도'], y = df['현역(징집)병 입영인원'], color = colors[3], label = '현역병 입영인원')
sns.lineplot(x = df['연도'], y = df['모집병 입영인원'], color = colors[0], label = '모집병 입영인원')
sns.lineplot(x = df['연도'], y = df['총합'], color = colors[2], label = '총 입영인원')
plt.legend()
plt.xlabel('연도')
plt.ylabel('입영 인원')
plt.title('연도별 징집병/모집병 입영인원', fontsize = 15)
plt.show()
```
